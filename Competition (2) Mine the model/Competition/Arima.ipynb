```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# List of CSV files
csv_files = csv_files = ['003.csv', '064.csv', '070.csv', '088.csv', '135.csv', '143.csv', '168.csv', '169.csv', '213.csv', '226.csv', '228.csv', '230.csv', '234.csv', '274.csv', '319.csv', '355.csv', '367.csv', '374.csv', '376.csv', '387.csv', '390.csv', '405.csv', '416.csv', '433.csv', '436.csv', '445.csv', '473.csv', '484.csv', '495.csv', '528.csv', '542.csv', '546.csv', '552.csv', '558.csv', '559.csv', '577.csv', '592.csv', '614.csv', '640.csv', '645.csv', '653.csv', '662.csv', '665.csv', '702.csv', '791.csv', '804.csv', '859.csv', '875.csv', '914.csv','958.csv']

# ARIMA hyperparameters
p, d, q = 5, 1, 2  # Example values for the order of AR, I, and MA components

# Number of days to forecast
forecast_steps = 100

for file_name in csv_files:
    # Load and preprocess the data for each CSV file
    data = pd.read_csv(f"./mine-the-model-2023/Upload-Dataset/TRAIN/{file_name}")
    stock_prices = data['Close']
    time_series = pd.Series(stock_prices)

    # Train an ARIMA model for each dataset
    model = ARIMA(time_series, order=(p, d, q))
    model_fit = model.fit()

    # Forecast the next 100 days
    forecast = model_fit.forecast(steps=forecast_steps)

    # Plot the original time series and the forecast for each file
    plt.figure()
    plt.plot(time_series, color='blue', label='Original Time Series')
    plt.plot(np.arange(len(time_series), len(time_series) + forecast_steps), forecast, color='red', label='Forecast')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.title(f'Stock Price Forecast for {file_name}')
    plt.legend()
    plt.show()






```
