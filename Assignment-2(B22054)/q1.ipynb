```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_csv("Iris.csv")
Y=np.array(d1['Species'])
Y



```

```python
d2=d1.drop(['Species'],axis=1)
attribute=d2.to_numpy()
attribute

```

# (b)

## Medianfill


```python
arr=attribute.T

arr
# col=d1.columns
# col=list(d1.columns)
for i in range(len(arr)):
    Q3=np.percentile(arr[i],75)
    Q1=np.percentile(arr[i],25)
    IQR=Q3-Q1
    for j in range(len(arr[i])):
        if arr[i][j]>(Q3+(1.5*IQR)) or arr[i][j]<(Q1-(1.5*IQR)):
            arr[i][j]== np.median(arr[i])
Original_Corrected = np.array(arr, copy=True)
arr.shape


```

# (C)

## PCA


```python
for i in range(len(arr)):
    arr[i]=arr[i]-np.mean(arr[i])
C=np.dot(arr,arr.T)/150
evalue,evector=np.linalg.eig(C)
evalue
evector = evector.T
evector
Top2 = evector[:2]
Top2
reduceddata=np.dot(Original_Corrected.T,Top2.T)
reduceddata




```

# (d)
## Scatter plot

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
rdata=scaler.fit_transform(reduceddata)
# rdata = reduceddata
evector
```

```python
plt.scatter(rdata.T[0],rdata.T[1],color='g',edgecolors='black')
plt.quiver(0, 0, evalue[0] * evector[0][0], evalue[0] * evector[0][1], angles='xy', scale_units='xy',color='red', scale=0.5, width=0.005)
plt.quiver(0, 0, evalue[1] * evector[1][0], evalue[1] * evector[1][1], angles='xy', scale_units='xy',color='red', scale=0.5, width=0.005, label='Eigen Vectors')
plt.grid()
plt.show()
```

# (e)
## Reconstructing original data

```python
newdata=np.dot(reduceddata,Top2)
newdata.shape
```

## KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
X_train,X_test,Y_train,Y_test =train_test_split(reduceddata,Y,random_state=104,test_size=0.20,shuffle=True)
```

```python

Y_test

```

```python
Euclidean=[]
for i in range(len(X_test)):

    l=[]

    for j in range(len(X_train)):

        distance= pow(np.sum(pow(X_test[i] - X_train[j] , 2)) , 0.5)
        l.append([distance, Y_train[j] ])

    l.sort(key = lambda x:x[0])
    Euclidean.append(l[:5])

Euclidean = np.array(Euclidean)
Euclidean
```

```python
import statistics as stat
ypredicted = np.array(   [  stat.mode(Euclidean[i][:,1]) for i in range(len(Euclidean))    ]   )
ypredicted
```

```python
f  = pd.DataFrame(np.array([Y_test,ypredicted]).T, columns=['Ypred' , 'YActual'])
f
```

```python
CM=confusion_matrix(Y_test,ypredicted,labels=["Iris-virginica","Iris-versicolor","Iris-setosa"])
CM
```

```python

ConfusionMatrixDisplay(CM).plot()
```
